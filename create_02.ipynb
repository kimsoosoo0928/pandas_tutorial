```python
import pandas as pd
import numpy as np
import random
import string
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(10))
```

```python
id()
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
score()
```

```python
pd.DataFrame(np.random.randint(0, 100, size=(10, 4)), index=['a','b','c','d','e','f','g','h','i','j'], columns=["국어","영어","수학","사회"])
```

```python
pd.DataFrame(np.random.randint(0, 100, size=(10, 4)), index=[id() for i in range(10)], columns=["국어","영어","수학","사회"])
```

```python
pd.DataFrame([[score() for i in range(1,5) ] for i in range(1,11)], index=[id() for i in range(1, 11)], columns=['국어','영어','수학','사회'])
```

```python
df = pd.DataFrame({"국어":score(), "영어":score(), "수학":score(), "사회":score()}, index=[id()])
for i in range(1,10):
    df.loc[id()] = {"국어":score(), "영어":score(), "수학":score(), "사회":score()}
df
```

```python
df.iloc[[0],:]
```

```python
df.loc[['Plkueojsop'],:]
```

```python
# 과학과목 추가, 기존 학생들에게 각 랜덤으로 추가하시오
df.loc[:,'과학'] = pd.Series([score() for i in range(10)], index=df.index)
df
```

```python
# 각 학생들의 점수의 총합을 표현하는 컬럽을 추가하세요
df.loc[:,'총점'] =  pd.Series(df.sum(1))
df
```

```python
df.sort_values('총점',ascending=False)
df
```
